```python
from glob import glob
```

```python
all_jpg = glob('./all_jpg/*')
```

```python
len(all_jpg)
```

```python
all_png = glob('./all_png/*')
```

```python
unique_jpg = [i.split('/')[-1][:-4] for i in all_jpg]
```

```python
unique_png = [i.split('/')[-1][:-4] for i in all_png]
```

```python
len(unique_png)
```

```python
delete_img = list(set(unique_png) - set(unique_jpg))
```

```python
files_to_delete_arr = ['/Users/bat/all_png/' + i + '.png' for i in delete_img]
```

```python
len(files_to_delete_arr)
```

```python
files_to_delete_arr
```

```python
import os
for i in files_to_delete_arr:
    try:
        os.remove(i)
    except:
        print(i)
print("File Removed!")
```
